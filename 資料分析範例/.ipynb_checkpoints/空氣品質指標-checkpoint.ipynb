## 空氣品質指標(AQI)
- 行政院環境保護署提供
- https://data.gov.tw/dataset/89029

```python
import pandas as pd
import requests
import csv

#CSV_URL = "https://data.epa.gov.tw/api/v1/aqx_p_432?limit=1000&api_key=9be7b239-557b-4c10-9775-78cadfc555e9&format=csv"
#下載並存檔
CSV_URL = "https://data.epa.gov.tw/api/v1/aqx_p_432?limit=1000&api_key=9be7b239-557b-4c10-9775-78cadfc555e9&format=csv"
response = requests.get(CSV_URL)
if response.status_code == 200:
    with open('空氣品質指標.csv','wb') as f:
        for chunk in response.iter_content(chunk_size=128):
            f.write(chunk)
else:
    print('下載錯誤')



#查詢DataFrame的相關資料
df = pd.read_csv('空氣品質指標.csv')
df.head(20)
```

```python
#依County欄位排序
pd.options.display.max_rows = 100
sorted_df=df.sort_values(by='County')
```

```python
sorted_df
```

```python
final_df = sorted_df.loc[:,['County', 'SiteName', 'AQI','PublishTime','PM2.5_AVG']]
```

```python
final_df.columns = ['縣市','位置','空氣品質','發佈時間','PM2.5']
```

```python
final_df
```

```python
#使用map建立一個空氣狀況

def status(x):
    if x <= 50:
        return '良好'
    elif x <= 100:
        return '普通'
    elif x <= 150:
        return '對敏感族群不健康'
    elif x <= 200:
        return '對所有族群不健康'
    elif x <= 300:
        return '非常不健康'
    else: 
        return '危害'

aqi = final_df.loc[:,'空氣品質']
status = aqi.map(lambda x: status(x))
status
```

```python
df1 = pd.concat([final_df, status],axis=1)
```

```python
df1.columns = ['縣市','位置','空氣品質','發佈時間','PM2.5','空氣狀況']
```

```python
df1
```

```python
df1.index = df1['縣市']+df1['位置']
df1.index
```

```python
df2 = df1.drop(['縣市','位置'],axis=1)
df2
```

```python
df3 = df2['空氣品質']
```

```python
#import FontProperties
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8) 
```

```python
#使用前10筆,顯示各縣市AQI值
ax = df3[:10].plot(style='ko--')
ax.set_xticklabels(df3.index,fontproperties=font)
```

```python
stateFrame = df2.groupby('空氣狀況').size()
figure1, ax1 = plt.subplots()
stateFrame.plot(kind='bar', ax = ax1,color=['r','g','orange'])
ax1.set_xticklabels(stateFrame.index,fontproperties=font,rotation=0)
ax1.set_xlabel('空氣狀況',fontproperties=font)
for p in ax1.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax1.annotate('{:}'.format(height), (x+width/2, y+ height+1),fontproperties=font)
```

```python
stateFrame = df2.groupby('空氣狀況')
stateDict=dict(list(stateFrame))
for state,df in stateDict.items():
    print(state)
    print(df)
    print("========================")
```
