```python
import requests
url = "https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/json?page=0&size=1000"
response = requests.get(url)
if response.status_code == 200:
    json_data = response.json()
```

```python
import pandas as pd 
import numpy as np

#重新安排所需要的欄位
allRecords = pd.DataFrame(json_data,columns=['sna', 'tot', 'sbi', 'sarea', 'ar', 'bemp'])
#更改欄位名稱
allRecords.rename(columns={"sna":"站名",
                           "tot":"數量",
                           "sbi":"可借",
                           "bemp":"可還",
                          })
```
