```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(8,5))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax3.plot(np.random.randn(50).cumsum(),'k--')
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax3.plot(np.random.randn(50).cumsum(),'k--')
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(np.random.randn(30).cumsum(),'go--')
plt.show()
```

```python
data = np.random.randn(30).cumsum()
fig = plt.figure(figsize=(8,5))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(data, 'k--', label='Default')
ax1.plot(data,'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc="best")
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(np.random.randn(1000).cumsum())
ax1.set_xticks([0, 250, 500, 750, 1000])
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(np.random.randn(1000).cumsum())
ax1.set_xticklabels(['one', 'two', 'three','four', 'five','six','seven'],rotation=30)
ax1.set_title("My plot")
ax1.set_xlabel('stage')
plt.show()
```
