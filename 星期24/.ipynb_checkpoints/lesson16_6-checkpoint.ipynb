```python
import requests

jsonUrl = 'http://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/8ae4cabf6973990e0169947ed32454b9'
response = requests.get(jsonUrl)
source_json = response.json()
```

```python
import pandas as pd
frame = pd.DataFrame(source_json)
frame
```

```python
staCode_url = 'http://ods.railway.gov.tw/tra-ods-web/ods/download/dataResource/0518b833e8964d53bfea3f7691aea0ee'
response = requests.get(staCode_url)
staCode_json = response.json()
```

```python
staCode_frame = pd.DataFrame(staCode_json,columns=['stationCode', 'stationName','stationAddrTw'])
staCode_frame
```

```python
merge_result = pd.merge(frame,staCode_frame,left_on='staCode',right_on='stationCode',how='left')
merge_result
```

```python
merge_result = merge_result[['trnOpDate','stationName','stationAddrTw','gateInComingCnt','gateOutGoingCnt']]
merge_result
```

```python
merge_result.columns = ['日期','站名','站址','進站人數','出站人數']
merge_result
```
