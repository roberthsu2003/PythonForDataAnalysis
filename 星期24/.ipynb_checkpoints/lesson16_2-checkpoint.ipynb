```python
import numpy as np
import pandas as pd

data = pd.Series(np.random.randn(9),
                index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],
                       [1, 2, 3, 1, 2, 3, 1, 2, 3]])
#透過階層式的索引,stack,unstack,建立不同的pivot
data
```

```python
data.unstack()
```

```python
data.unstack(0)
```

```python
#DataFrame使用多階層索引
frame = pd.DataFrame(np.arange(12).reshape((4, 3)),
                    index=[['a','a','b','b'],[1, 2, 1, 2]],
                    columns=[['台北','台北','台中'],['Green','Red','Green']])
frame
```

```python
frame.index.names = ['key1','key2']
frame.columns.names = ['縣市','顏色']
frame
```

```python
frame['台北']
```

```python
stackData=frame.stack().stack()
stackData
```

```python
stackData.unstack(level='key1')
```

```python
stackData.unstack(level=['縣市', 'key1'])
```

```python
#透過階層式的索引,goupby(level=xxxx)運算
frame
```

```python
frame.groupby(level='key1').sum()
```

```python
frame
```

```python
frame.groupby(level='顏色',axis=1).sum()
```

```python
frame.groupby(level='縣市',axis=1).sum()
```

```python
frame.reindex(columns=['台中','台北'],level='縣市')
```

```python
#只有單1個索引,如何改變成為階層式索引
frame = pd.DataFrame({'a':range(7), 'b':range(7, 0, -1),
                     'c':['one', 'one', 'one', 'two', 'two', 'two', 'two'],
                     'd':[0, 1, 2, 0, 1, 2, 3]})
frame
```

```python
#將column改變成為索引
#set_index()
frame.set_index('c')
```

```python
#set_index()
#將欄位成為階層式索引
newFrame = frame.set_index(['c','d'])
newFrame
```

```python
newFrame.reset_index()
```
