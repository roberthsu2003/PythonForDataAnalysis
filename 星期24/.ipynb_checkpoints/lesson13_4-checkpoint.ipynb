```python
import csv
import pandas as pd

with open('新北市食品工廠清冊.csv','r',encoding='utf-8') as file:    
    rows = csv.reader(file)
    datalist = [row for row in rows]
    title=datalist.pop(0)

dataFrame = pd.DataFrame(datalist,columns=title)
dataFrame    
```

```python
import requests
url = 'https://data.taipei/api/dataset/a0f33c38-bd7d-44bc-a33d-54a332ffd8f1/resource/98ae9fc8-092c-4898-9953-7758bf3784b7/download'
response = requests.get(url,stream=True)
with open('a.csv', 'wb') as fd:
    for chunk in response.iter_content(chunk_size=128):
        fd.write(chunk)
```

```python
with open('a.csv','r',encoding='big5') as file:    
    rows = csv.reader(file)
    datalist = [row for row in rows]
    title=datalist.pop(0)

dataFrame = pd.DataFrame(datalist,columns=title)
dataFrame 
```
