```python
import requests
url = "https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/json?page=0&size=1000"
response = requests.get(url)
if response.status_code == 200:
    json_data = response.json()
```

```python
import pandas as pd 
import numpy as np

#重新安排所需要的欄位
allRecords = pd.DataFrame(json_data,columns=['sna', 'tot', 'sbi', 'sarea', 'ar', 'bemp'])
#更改欄位名稱
                          
allRecords=allRecords.rename(columns={"sna":"站名",
                           "tot":"數量",
                           "sbi":"可借",
                           "bemp":"可還",
                           "sarea":"區域",
                           "ar":"地址"
                          })
allRecords
```

```python
#清理資料,全部有資料
allRecords.isnull().sum()
#全部0代表所有欄位有資料
```

```python
#轉換欄位資料
allRecords.dtypes
```

```python
allRecords['數量'] = allRecords['數量'].astype(int)
allRecords['可借'] = allRecords['可借'].astype(int)
allRecords['可還'] = allRecords['可還'].astype(int)
```

```python
allRecords.dtypes
```

```python
allRecords['故障'] = allRecords['數量'] - (allRecords['可借'] + allRecords['可還'])
allRecords
```

```python
#請取出目前雙和醫院的資料
allRecords[allRecords['站名']=='雙和醫院']
```

```python
#請取出目前雙和醫院的資料
allRecords.query('站名=="雙和醫院"')
```

```python
#請取出數量最多的站點
allRecords.sort_values(by='數量').iloc[-1:]
```

```python
#已經沒有車輛可借的站點
allRecords.query('可借==0')
```

```python
#所有區的總合資料
allRecords.groupby('區域')[['數量', '可借', '可還','故障']].sum()
```
