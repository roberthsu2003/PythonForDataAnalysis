```python
import csv


with open('新北市食品工廠清冊.csv','r',encoding='utf-8') as file:
    rows = csv.reader(file)
    listrows = list(rows)
    columns = listrows[0][1:]
    index = [item[0] for item in listrows[1:]]
    values = [ item[1:] for item in listrows[1:]]
print(columns)
```

```python
import pandas as pd
df = pd.DataFrame(values,index=index,columns=columns)
df
```
