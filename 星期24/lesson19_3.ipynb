```python
import numpy as np
import pandas as pd

xlsx = pd.ExcelFile('總統.xls')
all = pd.read_excel(xlsx,'中　央')
all
```

```python
df = all.iloc[5:,:4]
df.name = "第14任總統、副總統選舉候選人得票數一覽表"
df.columns = ['縣市', '朱立倫','蔡英文','宋楚瑜']
df
```

```python
df = df.reset_index().drop(columns='index')
df
```

```python
df = df.set_index('縣市')
df
```

```python
df.columns.name = "總統候選人"
df
```

```python
df = df[['朱立倫','蔡英文','宋楚瑜']].astype(int)
```

```python
df2 = df.sum()
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=10)
#font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=14)

figure = plt.figure(figsize=(8,5))
ax1 = figure.add_subplot(1,1,1)
df2.plot(ax=ax1, kind='bar', color=['b','g','orange'])
ax1.set_xticklabels(df2.index, fontproperties=font, rotation=30)
ax1.set_yticklabels(['0','100萬','200萬','300萬','400萬','500萬','600萬','700萬'],fontproperties=font)
ax1.set_title('2016得票數',fontproperties=font)
ax1.set_xlabel('總統候選人',fontproperties=font)
ax1.grid()
#ax1.legend(['朱立倫','蔡英文','宋楚瑜'],prop=font,loc ="best")
for p in ax1.patches:
    width, height = p.get_width(), p.get_height()
    x,y = p.get_xy()    
    print(width,height)
    print(x,y)
    ax1.annotate(f'{height}票',(x+width/4,y+height+100000),fontproperties=font)
plt.show()
```
