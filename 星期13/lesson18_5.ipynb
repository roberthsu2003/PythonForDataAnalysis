```python
import numpy as np
import pandas as pd

array = np.random.randn(10).cumsum()
print(array)

series = pd.Series(array,index=np.arange(0,100,10))
series
```

```python
series.plot()
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
ax1 = figure.add_subplot(1,1,1)
ax1.plot(series)
```

```python
df = pd.DataFrame(np.random.randn(10,4).cumsum(0),
                 index=np.arange(0,100,10),
                 columns = ['A','B','C','D']
                 )
figure = plt.figure()
ax1 = figure.add_subplot(1,1,1)
ax1.plot(df)
```

```python
df = pd.DataFrame(np.random.rand(6, 4),
                   index=['one', 'two', 'three', 'four', 'five', 'six'],
                   columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
df
```

```python
figure = plt.figure()
ax1 = figure.add_subplot(2,1,1)
ax2 = figure.add_subplot(2,1,2)
df.plot(ax=ax1,kind='bar')
df.plot(ax=ax2)
```

```python
df.plot.bar()
```

```python
df.plot.barh()
```

```python
df.plot.barh(stacked=True)
```

```python
df = pd.DataFrame(np.random.rand(6, 2),
                   index=['one', 'two', 'three', 'four', 'five', 'six'],
                   columns=pd.Index(['', ''], name='Genus'))
df
```
