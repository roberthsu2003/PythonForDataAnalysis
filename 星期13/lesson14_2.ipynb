```python
#存成csv檔
needsDataFrame.to_csv('needsDataFrame.csv')
```

```python
needsDataFrame.dtypes
```

```python
def toInt(x):
    try:
        return int(x)
    except:
        return 0
needsDataFrame['人口數'] = needsDataFrame['人口數'].map(toInt)
```

```python
def toFloat(x):
    try:
        return float(x)
    except:
        return 0.0
    
needsDataFrame['土地面積'] = needsDataFrame['土地面積'].map(toFloat)
```
