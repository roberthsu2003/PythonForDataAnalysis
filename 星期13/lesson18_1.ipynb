```python
import numpy as np
import pandas as pd

tips = pd.read_csv('tips.csv')
tips
```

```python
tips.columns = ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
tips.dtypes
```

```python
tips.info()
```

```python
tips['小費比例'] = tips['小費'] / tips['總票價']
tips
```

```python
#建立groupby物件,先確認要groupby的欄位
grouped = tips.groupby(['日期','吸煙者'])
grouped
```

```python
#確定值是那幾個欄位
grouped_pct = grouped['小費']#subscript語法
s1=grouped_pct.mean()
s1
```

```python
s1.unstack()
```

```python
grouped[['小費','小費比例']].mean()
```
