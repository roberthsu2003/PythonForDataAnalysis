```python
import numpy as np
import pandas as pd

xlsx = pd.ExcelFile('總統.xls')
```

```python
all=pd.read_excel(xlsx,'中　央')
all
```

```python
df = all.iloc[5:,:4]
df.name = '第14任總統、副總統選舉候選人得票數一覽表'
df.columns = ['縣市','朱立倫','蔡英文','宋楚瑜']
df
```

```python
#檢查資料
#null
df.isnull().sum()
df.info()
```

```python
#縣市成為索引
df=df.set_index('縣市')
df.columns.name = '總統候選人'
df
```

```python
def toInt(value):
    try:
        return int(value)
    except:
        print(value)

df['朱立倫'].map(toInt)
```

```python
def pd_to_int(s):
    convertList = [int(item) for item in s]
    return pd.Series(convertList,index=s.index)
    
df.apply(pd_to_int)
```

```python
df=df.astype('int32')
df.dtypes
```

```python
df2 = df.sum()
df2
```

```python
#bar圖形
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=14)
figure = plt.figure()
ax1 = figure.add_subplot(1,1,1)
df2.plot(ax = ax1,kind='bar',color=['b','g','orange'])
ax1.set_xticklabels(df2.index,fontproperties=font,rotation=45)
ax1.set_xlabel(df2.index.name,fontproperties=font)
ax1.set_yticklabels([0,1000000,2000000,3000000,4000000,5000000,6000000,7000000,8000000])

for p in ax1.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    print(x,y,width,height)    
    newX,newY = x+0.1 , y+height-500000
    ax1.annotate('{:}票'.format(height),(newX,newY),fontproperties=font,color='w')
```

```python
figure1 = plt.figure()
ax2 = figure1.add_subplot(1,1,1)
df.plot(ax = ax2, kind='bar', color=['b','g','orange'])
ax2.set_xticklabels(df.index,fontproperties=font)
ax2.legend(prop=font)
ax2.set_xlabel('縣市',fontproperties=font)
ax2.set_title('第14任總統、副總統選舉候選人得票數一覽表',fontproperties=font)
ax2.annotate('自訂的文字',(1,1000000),fontproperties=font)
```
