```python
import matplotlib.pyplot as plt
data = [-1, -4.3, 15, 21, 31]
plt.figure(num=3,figsize=(8,5))
plt.plot(data)
plt.show()
```

```python
x = ['chinese', 'english', 'math', 'history', 'discover']
y = [89,70,69,76,84]
y1 = [70,89,92,76,68]

plt.figure(num=4, figsize=(8,5))
plt.plot(x, y,'r--o', x, y1,'bo')
plt.show()
```

```python
x = [1, 2,3,4,5]
y1 = [1,2,5,6,7]
y2 = [4, 5, 6, 7, 8]
plt.figure()
plt.plot(x, y1, "r-o", x, y2, "g--")
plt.grid()
plt.axis([1.8,4.2,1,7])
plt.show()
```

```python
import random
plt.figure(dpi=300)
days = range(1,31)
celsius = [random.randint(0,35) for day in range(1,31)]
plt.figure()
plt.plot(days,celsius,'r-o')
plt.savefig('test.png',dpi=300)
plt.show()
```
