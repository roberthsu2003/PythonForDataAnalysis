```python
import pandas_datareader.data as web
web.get_data_yahoo('2330.TW')['Adj Close']
```

```python
all_data = {ticker:web.get_data_yahoo(ticker)['Adj Close'] for ticker in ['2330.TW','2303.TW','2454.TW','2317.TW']}
all_data
```

```python
import numpy as np
import pandas as pd
price = pd.DataFrame(all_data)
price
```

```python
price.columns = ['台積電','聯電','聯發科','鴻海']
```

```python
price.loc['2022-01-01':'2022-04-18']
```

```python
pct_price=price.pct_change().dropna()
pct_price
```

```python
pct_price['台積電'].corr(pct_price['聯電'])
```
