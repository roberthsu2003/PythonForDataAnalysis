```python
import csv

with open('各鄉鎮市區人口密度.csv','r',encoding='utf-8') as file:
    next(file)
    reader = csv.reader(file)
    data = []
    for item in reader:
        if(item[0] == '106' or item[0]=='統計年'):
            data.append(item)
```

```python
import pandas as pd
titles = data.pop(0)
```

```python
dataFrame=pd.DataFrame(data,columns=titles)
dataFrame[50:101]
```
