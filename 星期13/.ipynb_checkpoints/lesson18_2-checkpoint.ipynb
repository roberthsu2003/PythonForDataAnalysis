```python
import numpy as np
import pandas as pd

tips = pd.read_csv('tips.csv')
tips.columns = ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
tips['小費比例'] = tips['小費'] / tips['總票價']
tips
```

```python
tips.sort_values(by='小費比例',ascending=False)[:5]
```

```python
def top(df,n=5,column='小費比例'):
    return df.sort_values(by=column,ascending=False)[:n]

top(tips,n=3,column='總票價')
```

```python
top(tips)
```

```python
tips.groupby('吸煙者').apply(top)
```

```python
tips.groupby('吸煙者').apply(top,n=3)
```
