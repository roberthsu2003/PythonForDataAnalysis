```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3,num=50)
y1 = 2*x + 1
y2 = x ** 2

plt.figure()
plt.plot(x,y1,color='red',linewidth=1.0, linestyle='--')
plt.plot(x,y2,'b-')
plt.title('numpy test')
plt.xlabel('Day')
plt.ylabel('Celsius')
plt.show()
```
