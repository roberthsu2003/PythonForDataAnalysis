```python
import matplotlib.pyplot as plt
import numpy as np

x = ['apple', 'banana', 'orange', 'tomato', 'guava']
y = [5, 7, 3, 4, 6]
plt.figure()
plt.bar(x, y)
plt.show()
```

```python
x = ['apple', 'banana', 'orange', 'tomato', 'guava']
y = [5, 7, 3, 4, 6]
plt.figure()
plt.barh(x, y)
plt.show()
```

```python
#修改預設的xticks
index = np.arange(5)
value = np.array([5, 7, 3, 4, 6]) * 1000
plt.figure()
plt.bar(index,value)
plt.xticks(index,['apple', 'banana', 'orange', 'tomato', 'guava'])
#plt.yticks(value,['1*1000','2*1000','3*1000','4*1000','5*1000'])
plt.show()
```

```python
labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [10, 30, 45, 15]
colors = ['yellow', 'green', 'red', 'blue']
plt.figure()
plt.pie(values,labels=labels, colors=colors)
plt.show()
```
