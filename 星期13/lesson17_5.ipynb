```python
import numpy as np
import pandas as pd

tips = pd.read_csv('tips.csv')
tips
```

```python
tips.columns = ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
tips
```

```python
tips['小費比例'] = tips['小費']/tips['總票價'] 
```

```python
tips
```

```python
grouped = tips.groupby(['日期','吸煙者'])
grouped_pct = tips.groupby(['日期','吸煙者'])['小費']
grouped_pct.mean()
```

```python
grouped_pct.agg(['mean','std'])
```
