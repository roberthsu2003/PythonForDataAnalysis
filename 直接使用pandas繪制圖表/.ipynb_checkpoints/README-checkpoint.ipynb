## Figures and subplots 

```python
#一個figure,4個subplot
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) # 2列, 2欄, 第1個圖表
ax2 = fig.add_subplot(2,2,2) # 2列, 2欄, 第2個圖表
ax3 = fig.add_subplot(2,2,3) # 2列, 2欄, 第3個圖表
#ax1是AxesSubPlot物件
#會畫在最後一個ax3
plt.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
#一個figure,4個subplot
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) # 2列, 2欄, 第1個圖表
ax2 = fig.add_subplot(2,2,2) # 2列, 2欄, 第2個圖表
ax3 = fig.add_subplot(2,2,3) # 2列, 2欄, 第3個圖表
ax1.hist(np.random.randn(100),bins=20,color='k', alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30) + 3 * np.random.randn(30))
ax3.plot(np.random.randn(50).cumsum(), 'k--')
```

### Ticks,Labels,Legends
- xlim, xticks, xticklabels

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
```

```python
#set_xticks()
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five',c'six',c'seven'], rotation=30, fontsize='small')
ax.set_title('My First matplotlib plot')
ax.set_xlabel('Stage')
```

```python
#增加legends
from numpy.random import randn
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(), 'k', label='one')
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc='best')
```

### Annotations and Drawing on a Subplot

ax.text(x, y, 'Hello world!',
            family='monospace', fontsize=10)

```python
from datetime import datetime
import pandas as pd

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
data = pd.read_csv('spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']
spx.plot(ax=ax, style='k-')
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                        xytext=(date, spx.asof(date) + 225),
                        arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                        headlength=4),
                        horizontalalignment='left', verticalalignment='top')
# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='g', alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

## Plotting with pandas and seaborn

### Line Plots


```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10))
df.plot()
```

### Bar Plots

```python
fig, axes = plt.subplots(2, 1)
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0], color='k', alpha=0.7)
data.plot.barh(ax=axes[1], color='k', alpha=0.7)
```

```python
import pandas as pd
tips = pd.read_csv('tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts = party_counts.loc[:, 2:5]
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
party_pcts
```

```python
party_pcts.plot.bar()
```
