## Figures and subplots 

````python
#一個figure,4個subplot
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) # 2列, 2欄, 第1個圖表
ax2 = fig.add_subplot(2,2,2) # 2列, 2欄, 第2個圖表
ax3 = fig.add_subplot(2,2,3) # 2列, 2欄, 第3個圖表
#ax1是AxesSubPlot物件
#會畫在最後一個ax3
plt.plot(np.random.randn(50).cumsum(), 'k--')
````

````python
#一個figure,4個subplot
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) # 2列, 2欄, 第1個圖表
ax2 = fig.add_subplot(2,2,2) # 2列, 2欄, 第2個圖表
ax3 = fig.add_subplot(2,2,3) # 2列, 2欄, 第3個圖表
ax1.hist(np.random.randn(100),bins=20,color='k', alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30) + 3 * np.random.randn(30))
ax3.plot(np.random.randn(50).cumsum(), 'k--')
````

### 調整圖表的外部距離和各個圖表的間
- 可以使用figure實體
- 可以使用top function
```
subplots_adjust(left=None, bottom=None, right=None, top=None,
wspace=None, hspace=None)
```


````python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=50, color='k', alpha=0.5)
fig.subplots_adjust(wspace=0.1, hspace=0)
plt.show()
````

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
ax.set_xticks([0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]) #先使用set_xticks,再使用set_xticklabels
labels = ax.set_xticklabels(['zero','one', 'two', 'three', 'four', 'five','six','seven','eight','nine','ten'], rotation=30, fontsize='small')
ax.set_title('My First matplotlib plot')
ax.set_xlabel('Stage')
````

````python
#增加legends
from numpy.random import randn
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(), 'k', label='one')
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc='best')
````

### Annotations and Drawing on a Subplot
- text()可以畫文字和設定樣式
- annotation可以畫文字和箭頭

ax.text(x, y, 'Hello world!',
            family='monospace', fontsize=10)

````python
from datetime import datetime
import pandas as pd

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
data = pd.read_csv('spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']
spx.plot(ax=ax, style='k-')
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                        xytext=(date, spx.asof(date) + 225),
                        arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                        headlength=4),
                        horizontalalignment='left', verticalalignment='top')
# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2008-2009 financial crisis')
````

````python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='g', alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
````

````python
import pandas as pd
tips = pd.read_csv('tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts
````

````python
party_counts = party_counts.loc[:, 2:5]
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
party_pcts
````

````python
party_pcts.plot.bar()
````

````python

tips['tip_pct'] = tips['tip'] / (tips['total_bill']-tips['tip'])
tips.head()
````
