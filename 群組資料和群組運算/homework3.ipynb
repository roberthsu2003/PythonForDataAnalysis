```python
# 請處理(新北市公共自行車租賃系統.json)
# google搜尋online json viewer
#api接口:https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/json?page=0&size=1000

import urllib3
import certifi
import json
import pandas as pd

http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED',ca_certs=certifi.where())
r = http.request('GET', 'https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/json?page=0&size=1000')
if r.status == 200:
    records = json.loads(r.data.decode('utf-8'))   
    
else:
    print("下載失敗")

allRecords = pd.DataFrame(records,columns=['sna', 'tot', 'sbi', 'sarea', 'ar', 'bemp'])
allRecords=allRecords.rename(columns = {'sna':'站名', 'sarea':'區域', 'ar':'地址', 'tot':'數量','sbi':'可借', 'bemp':'可還'})
allRecords
```

```python
allRecords.dtypes
```

```python
allRecords[['數量','可借','可還']]=allRecords[['數量','可借','可還']].astype(int)
allRecords.dtypes
```

```python
allRecords['故障'] = allRecords['數量'] - (allRecords['可借'] + allRecords['可還'])
allRecords['故障率'] = allRecords['故障'] / allRecords['數量']
allRecords
```

```python
grouped = allRecords.groupby('區域')
result = grouped[['數量','可還', '可借','故障']].agg([('總量','sum')])
result
```

```python
result['故障率'] = (result['故障']/result['數量'])
```

```python
result['故障率'] = result['故障率'].map(lambda x:'{:.2f}%'.format(x*100))
```

```python
result.columns = ['數量','可還','可借','故障','故障率']
result
```
