```python
#這是測試
import numpy as np
import pandas as pd
dataFrame = pd.read_csv('world.csv')
dataFrame
```

```python
dataFrame.info()
```

```python
#'洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數'
allDataFrame=dataFrame.reindex(columns=['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數'])
allDataFrame
```

```python
#請找出台灣資料
#[]搜尋
#query()搜尋
taiwan=allDataFrame.query('國家=="台灣"')
taiwan
```

```python
#請找出台灣資料,新增確診數0
#[]搜尋
#query()搜尋
taiwan1=allDataFrame.query('國家=="台灣" and 新增確診數==0')
taiwan1
```

```python
#請找出台灣資料,新增確診數>=10000
#[]搜尋
#query()搜尋
taiwan1=allDataFrame.query('國家=="台灣" and 新增確診數>=10000')
taiwan1
```

```python
#請找出台灣資料,最多新增確診數的資料
#[]搜尋
#query()搜尋
allDataFrame.query('國家=="台灣"')['新增確診數'].max()
```

```python
#請找出台灣資料,最多新增確診數的日期
#[]搜尋
#query()搜尋
max = allDataFrame.query('國家=="台灣"')['新增確診數'].max()
allDataFrame.query('新增確診數==@max')
```

```python
#請找出日本資料,最多新增確診數的日期
#[]搜尋
#query()搜尋
max = allDataFrame.query('國家=="日本"')['新增確診數'].max()
allDataFrame.query('新增確診數==@max')
```

```python
#請找出美國資料,最多新增確診數的日期
#[]搜尋
#query()搜尋
max = allDataFrame.query('國家=="美國"')['新增確診數'].max()
allDataFrame.query('新增確診數==@max')
```

```python
#請找出台灣資料,日期2022年的資料
#[]搜尋
#query()搜尋
allDataFrame.query('國家=="台灣" and 日期>="2022-01-01"')
```
