### 世界covid19.csv資料


```python
import pandas as pd
import numpy as np

dataFrame = pd.read_csv('world.csv')
dataFrame.head()
```

```python
allDataFrame = dataFrame.reindex(columns=['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數'])
allDataFrame
```

```python
taiwanFrame = allDataFrame[allDataFrame['國家'] == '台灣']
taiwanFrame
```

```python
#將使用pd.to_datatime()日期轉換為datatime
#有使用&一定要用()號
taiwan2022 = taiwanFrame[(taiwanFrame['日期']>'2022-01-01') & (taiwanFrame['日期']<='2022-07-06')]
taiwan2022
```

```python
taiwan2022 = taiwanFrame.query('日期 > "2022-01-01"' and '日期 <= "2022-07-06"')
taiwan2022
```

```python
japanFrame = allDataFrame[allDataFrame['國家'] == '日本']
japanFrame
```

```python
japan2022 = japanFrame[(japanFrame['日期']>'2022-01-01') & (japanFrame['日期']<='2022-07-06')]
japan2022
```

```python
def call(s):
    return pd.Series([s.min(), s.max(), s.mean()], index=['最少','最多','中間值'])

taiwan = pd.DataFrame(taiwan2022['新增確診數'].values,columns=['台灣'])
japan = pd.DataFrame(japan2022['新增確診數'].values,columns=['日本'])
taiwan1 = taiwan.apply(call)
japan1 = japan.apply(call)
taiwan1.add(japan1,fill_value=0)
```
