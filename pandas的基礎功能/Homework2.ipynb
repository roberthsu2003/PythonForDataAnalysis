### 世界covid19.csv資料


```python
import pandas as pd
import numpy as np

dataFrame = pd.read_csv('world.csv')
dataFrame.head()
```

```python
allDataFrame = dataFrame.reindex(columns=['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數'])
allDataFrame
```

```python
taiwanFrame = allDataFrame[allDataFrame['國家'] == '台灣']
taiwanFrame
```

```python
#將使用pd.to_datatime()日期轉換為datatime
#有使用&一定要用()號
taiwanFrame[(taiwanFrame['日期']>'2022-01-01') & (taiwanFrame['日期']<='2022-07-06')]
```

```python
taiwanFrame.query('日期 > "2022-01-01"' and '日期 <= "2022-07-06"')
```
