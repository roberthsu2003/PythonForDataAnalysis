### 將各鄉鎮市區人口密度修改成為理想的樞紐分析表
![結果](d1.png)

```python
import pandas as pd
import numpy as np

dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame
```

```python
#使用dtypes屬性查詢dataFrame所有的欄的資料類型
dataFrame.dtypes
```

```python
#使用shape屬性查詢欄列數量
dataFrame.shape
```

```python
#改變欄位索引標籤名稱
dataFrame.columns = dataFrame.loc[0]
dataFrame
```

```python
#查詢列索引
dataFrame.index
```

```python
#移除第一列
dataFrame = dataFrame.drop(0)
dataFrame
```

```python
#查看最後10行
dataFrame.iloc[-10:]
```

```python
#使用drop刪除最後幾行
dataFrame.drop([371, 372, 373, 374, 375],inplace=True)
dataFrame
```

```python
#建立新的索引列
#先用shape取出列的數量
rowCount, columnCount = dataFrame.shape
rowCount
```

```python
#建立新的索引列
dataFrame.index = range(rowCount)
dataFrame
```

```python
#使用reindex取出想要的欄位
dataFrame1 = dataFrame.reindex(columns=['區域別','年底人口數','土地面積'])
dataFrame1
```

```python
#使用rename()實體方法改變欄位名稱
dataFrame1 = dataFrame1.rename(columns={'年底人口數':'人口數'})
dataFrame1
```
