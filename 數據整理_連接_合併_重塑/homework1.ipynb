```python
#homework
#讀取ex6.csv
#將key欄位設定成為row的index
import numpy as np
import pandas as pd
result = pd.read_csv('ex6.csv')
result.index.name='編號' #改變row索引名稱
result
```

```python
#保留原來的索引,建立row的階層式索引
newResult = result.reset_index().set_index(['key', '編號']) 
newResult
```

```python
#取出欄位one
newSerial = newResult['one']
newSerial
```

```python
#排序索引key
newSerial.sort_index(level='key',inplace=True)
newSerial.index
```

```python

#建立樞紐分析表
newSerial.unstack()
```

```python
#依照索引key做欄位加總
newSerial.groupby('key').sum()
```

```python
#將newResult成為Serial
newResult.stack()
```

```python
#加總相同row索引key的值
newResult.groupby(level='key').sum()
```

```python
#加總各欄位
newResult.sum(axis=0)
```

```python
#取出各欄位加總
newResult['one'].sum()
newResult['two'].sum()
newResult['three'].sum()
newResult['four'].sum()
```
