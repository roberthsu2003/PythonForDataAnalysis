```python
import csv
import numpy as np
import pandas as pd 

with open('新北市食品工廠清冊.csv','r',encoding='utf-8') as file:
    next(file)
    rows = csv.reader(file)    
    dict = {row[1]:row[0] for row in rows}

stores = pd.Series(dict)
stores.index.name='統一編號'
stores.name = '工廠統編清冊'
stores
```

```python
#pd.Series.sort_index(stores)
stores=stores.sort_index(ascending=True)
stores
```
