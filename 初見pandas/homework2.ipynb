```python
import pandas as pd
import numpy as np
import csv

with open('新北市食品工廠清冊.csv','r',encoding='utf-8') as file:
    rows = csv.reader(file)    
    dataList = list(rows)
    titles = dataList.pop(0) #抓出第一列
    
    
dataFrame = pd.DataFrame(dataList,columns=titles)
dataFrame.columns = ['工廠','工廠登記編號','地址','統編','97x','97y', '緯度', '經度']
result_df= dataFrame[['工廠','工廠登記編號','地址','統編', '緯度', '經度']]
result_df
```
