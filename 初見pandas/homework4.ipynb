```python
# 請處理(新北市公共自行車租賃系統.json)
# google搜尋online json viewer

import pandas as pd
import numpy as np
import json
with open('新北市公共自行車租賃系統.json') as json_file:
    data = json.load(json_file)
    if data['success']:
        records = data['result']['records']
    else:
        print('讀取失敗')
dataFrame = pd.DataFrame(records,columns=['sno', 'sarea', 'sna', 'ar', 'tot', 'sbi', 'bemp', 'lat', 'lng'])
dataFrame.columns = ['編號', '區域', '站名', '地址', '全部數量', '可借', '可還', '緯度', '經度']
dataFrame
```

```python
dataFrame=dataFrame[dataFrame['區域'] == '汐止區']
dataFrame
```

```python
all=dataFrame['全部數量'].astype(np.int64)
print('可停車總數是為:',all.sum())
```

```python
sbi = dataFrame['可借'].astype(np.int64)
print('可借車總數為:',sbi.sum())
```

```python
bemp = dataFrame['可還'].astype(np.int64)
print('可還車總數為:',bemp.sum())
```
