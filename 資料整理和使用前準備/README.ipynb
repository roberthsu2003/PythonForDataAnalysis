### 清除NaN的資料

```python
# 清除NaN的資料
from numpy import nan as NA
data = pd.Series([1, NA, 3.5, NA, 7])
data
'''
Out[6]: 
0    1.0
1    NaN
2    3.5
3    NaN
4    7.0
dtype: float64
'''
#使用dropna()方法,清除遺失資料
data.dropna()
'''
Out[7]: 
0    1.0
2    3.5
4    7.0
dtype: float64
'''

#使用過濾方式方法,清除遺失資料
data[data.notnull()]
'''
Out[8]: 
0    1.0
2    3.5
4    7.0
dtype: float64
'''
```

```python
#DataFrame的清除比較複雜一些
#drapna()將會清除整個row
data = pd.DataFrame([[1., 6.5, 3.],
                    [1., NA, NA],
                    [NA, NA, NA],
                    [NA, 6.5, 3.]])
data
'''
0  1.0  6.5  3.0
1  1.0  NaN  NaN
2  NaN  NaN  NaN
3  NaN  6.5  3.0
'''

data.dropna()
'''
Out[10]: 
     0    1    2
0  1.0  6.5  3.0
'''

#做用引數名稱how='all',只清除整列都是NaN的列
data.dropna(how='all')
'''
Out[11]: 
     0    1    2
0  1.0  6.5  3.0
1  1.0  NaN  NaN
3  NaN  6.5  3.0
'''

#刪除整欄資料使用axis=1
data[4] = NA
data
'''
Out[12]: 
     0    1    2   4
0  1.0  6.5  3.0 NaN
1  1.0  NaN  NaN NaN
2  NaN  NaN  NaN NaN
3  NaN  6.5  3.0 NaN
'''

data.dropna(axis=1, how='all')
'''
Out[14]: 
     0    1    2
0  1.0  6.5  3.0
1  1.0  NaN  NaN
2  NaN  NaN  NaN
3  NaN  6.5  3.0
'''

df = pd.DataFrame(np.random.randn(7, 3))
df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA
df
'''
Out[15]: 
          0         1         2
0 -0.384156       NaN       NaN
1  2.457934       NaN       NaN
2  0.387037       NaN  1.516058
3  0.799194       NaN  1.019982
4 -0.523685  2.448814 -1.618673
5  0.139726 -0.483431  1.087822
6 -0.333721 -0.491657 -0.834928
'''

df.dropna()
'''
Out[16]: 
          0         1         2
4 -0.523685  2.448814 -1.618673
5  0.139726 -0.483431  1.087822
6 -0.333721 -0.491657 -0.834928
'''

#有2個的才刪除row,使用引數名稱thresh=2
df.dropna(thresh=2)
'''
Out[17]: 
          0         1         2
2  0.387037       NaN  1.516058
3  0.799194       NaN  1.019982
4 -0.523685  2.448814 -1.618673
5  0.139726 -0.483431  1.087822
6 -0.333721 -0.491657 -0.834928
'''
```

### 移除重複資料
- duplicated

```python
#duplicated 回傳boolean Series
import numpy as np
import pandas as pd

data = pd.DataFrame({'k1':['one', 'two']*3 + ['two'],
                    'k2':[1, 1, 2, 3, 3, 4, 4]})
data
'''
Out[26]: 
    k1  k2
0  one   1
1  two   1
2  one   2
3  two   3
4  one   3
5  two   4
6  two   4
'''
data.duplicates()

'''
Out[4]: 
0    False
1    False
2    False
3    False
4    False
5    False
6     True
dtype: bool

'''

#drop_duplicates()移除重覆資料
data.drop_duplicates()
'''
Out[4]: 
    k1  k2
0  one   1
1  two   1
2  one   2
3  two   3
4  one   3
5  two   4
'''

#移除欄位有重覆資料的列
data['v1'] = range(7)
data.drop_duplicates(['k1'])
'''
Out[5]: 
    k1  k2
0  one   1
1  two   1
'''

#drop_duplicates()保留第一筆,使用引數名稱keep='last',何留最後一筆
data.drop_duplicates(['k1', 'k2'], keep='last')
'''
Out[6]: 
    k1  k2
0  one   1
1  two   1
2  one   2
3  two   3
4  one   3
6  two   4
'''
```

### *使用function或map轉換資料

```python
data = pd.DataFrame({'食物':['培根','豬肉','培根','熏牛肉','咸牛肉','培根','熏牛肉','蜂蜜火腿','熏鮭魚'],
                    '重量':[4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
'''
Out[7]: 
     食物    重量
0    培根   4.0
1    豬肉   3.0
2    培根  12.0
3   熏牛肉   6.0
4   咸牛肉   7.5
5    培根   8.0
6   熏牛肉   3.0
7  蜂蜜火腿   5.0
8   熏鮭魚   6.0
'''

#使用Series.map({})方法,並且傳入對應的dictionary,建立有對應資料新的Series
mean_to_animal = {
   '培根':'豬',
    '豬肉':'豬',
    '熏牛肉':'牛',
    '咸牛肉':'牛',
    '蜂蜜火腿':'豬',
    '熏鮭魚':'鮭魚'
}

data['動物'] = data['食物'].map(mean_to_animal)
data
'''
Out[8]: 
     食物    重量   動物
0    培根   4.0    豬
1    豬肉   3.0    豬
2    培根  12.0    豬
3   熏牛肉   6.0    牛
4   咸牛肉   7.5    牛
5    培根   8.0    豬
6   熏牛肉   3.0    牛
7  蜂蜜火腿   5.0    豬
8   熏鮭魚   6.0   鮭魚
'''

#也可以使用Series.map(lambda)
data['食物'].map(lambda x: mean_to_animal[x])
'''
Out[9]: 
0      豬
1      豬
2      豬
3      牛
4      牛
5      豬
6      牛
7      豬
8     鮭魚
Name: 食物, dtype: object
'''
```

### *重新命名軸的索引名稱
- 使用function或mapping

```python
data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                   index=['台北','台中','台南'],
                   columns = ['one', 'two', 'three', 'four'])
data
'''
Out[15]: 
    one  two  three  four
台北    0    1      2     3
台中    4    5      6     7
台南    8    9     10    11
'''

#Series的index也有map()方法
transform = lambda x: x + '(台灣)'
data.index.map(transform)
#Out[16]: Index(['台北(台灣)', '台中(台灣)', '台南(台灣)'], dtype='object')

data.index = data.index.map(transform)
data
'''
Out[17]: 
        one  two  three  four
台北(台灣)    0    1      2     3
台中(台灣)    4    5      6     7
台南(台灣)    8    9     10    11
'''
#使用rename(),複制一個新的並重新命名
data.rename(index=str, columns=str.upper)
'''
Out[18]: 
        ONE  TWO  THREE  FOUR
台北(台灣)    0    1      2     3
台中(台灣)    4    5      6     7
台南(台灣)    8    9     10    11
'''

#使用rename(),並應用對應值來置換
data.rename(
    index={'台南(台灣)':'高雄(台灣)'},
    columns = {'one':'NUM'}
)
'''
Out[19]: 
        NUM  two  three  four
台北(台灣)    0    1      2     3
台中(台灣)    4    5      6     7
高雄(台灣)    8    9     10    11
'''

#使用引數名稱inplace=True,修改原始內容
data.rename(index={'台南(台灣)':'高雄(台灣)'}, inplace=True)
data
'''
Out[20]: 
        one  two  three  four
台北(台灣)    0    1      2     3
台中(台灣)    4    5      6     7
高雄(台灣)    8    9     10    11
'''
```

## 處理字串類型

### python字串物件的方法

```python
val = 'a,b,  guido'
val.split(',')
#Out[14]: ['a', 'b', '  guido']

#使用strip()方法,清除字串
pieces = [x.strip() for x in val.split(',')]
pieces
#Out[15]: ['a', 'b', 'guido']

#拆解list
first, second, third = pieces
first + '::' + second + '::' + third
#Out[16]: 'a::b::guido'

#使用join()
'::'.join(pieces)
#Out[17]: 'a::b::guido'

#使用index()和find()方法
'guido' in val
#Out[18]: True

val.index(',')
#Out[19]: 1

val.find(':')
#Out[20]: -1

val.count(',')
#Out[21]: 2

val.replace(',', '::')
#Out[22]: 'a::b::  guido'

val.replace(',', '')
Out[23]: 'ab  guido'
```

 - 替換(substitution)
 - 切割(splitting)
 

```python
import re
text = 'foo      bar\t  baz   \tqux'
re.split('\s+', text)
#Out[3]: ['foo', 'bar', 'baz', 'qux']

regex = re.compile('\s+')
regex.split(text)
#Out[4]: ['foo', 'bar', 'baz', 'qux']

regex.findall(text)
#Out[5]: [' ', '\t ', ' \t']

#findall(),搜尋所有字串,傳出符合的list
text = """
Dave dave@google.com 
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com
"""

pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'

#re.IGNORECASE
regex = re.compile(pattern, flags=re.IGNORECASE)
regex.findall(text)
#Out[6]: ['dave@google.com', 'steve@gmail.com', 'rob@gmail.com', 'ryan@yahoo.com']

#search() 搜尋所有字串,傳出第一個符合的字串,但是是Mach物件
m = regex.search(text)
text[m.start():m.end()]
#Out[7]: 'dave@google.com'

#match搜尋文字的起始處，沒發現則傳出None
print(regex.match(text))
#None

#sub(),傳出新字串
print(regex.sub('READCTED', text))
'''
Dave READCTED 
Steve READCTED
Rob READCTED
Ryan READCTED
'''

pattern = r'([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})'

regex = re.compile(pattern, flags=re.IGNORECASE)
m = regex.match('robert@gmail.com')
m.groups()
#Out[10]: ('robert', 'gmail', 'com')

regex.findall(text)
'''
Out[11]: 
[('dave', 'google', 'com'),
 ('steve', 'gmail', 'com'),
 ('rob', 'gmail', 'com'),
 ('ryan', 'yahoo', 'com')]
'''

print(regex.sub(r'Username: \1, Domain: \2, Suffix:\3',text))
'''
Dave Username: dave, Domain: google, Suffix:com 
Steve Username: steve, Domain: gmail, Suffix:com
Rob Username: rob, Domain: gmail, Suffix:com
Ryan Username: ryan, Domain: yahoo, Suffix:com
'''
```

### 使用pandas內建字串功能

```python
import numpy as np
import pandas as pd

data = {'Dave':'dave@google.com', 'Steve':'steve@gmail.com', 'Rob':'rob@gmail.com', 'Wes':np.nan}
data = pd.Series(data)
data
'''
Out[13]: 
Dave     dave@google.com
Steve    steve@gmail.com
Rob        rob@gmail.com
Wes                  NaN
dtype: object
'''

data.isnull()
'''
Out[14]: 
Dave     False
Steve    False
Rob      False
Wes       True
dtype: bool
'''

#使用str.contains()
data.str.contains('gmail')
'''
Out[15]: 
Dave     False
Steve     True
Rob       True
Wes        NaN
dtype: object
'''

#也可以使用正規則字串格式
pattern = '([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\\.([A-Z]{2,4})'
data.str.findall(pattern, flags=re.IGNORECASE)

'''
Out[16]: 
Dave     [(dave, google, com)]
Steve    [(steve, gmail, com)]
Rob        [(rob, gmail, com)]
Wes                        NaN
dtype: object
'''

matches = data.str.match(pattern, flags=re.IGNORECASE)
matches
'''
Out[17]: 
Dave     True
Steve    True
Rob      True
Wes       NaN
dtype: object
'''

```
