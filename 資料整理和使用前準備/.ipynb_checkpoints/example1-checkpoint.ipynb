```python
dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame[-20:]
```

```python
#有部份一列只要一個NaN
#有部份一列有多個NaN
#有部份的列有字串...
dataFrame.dropna() #刪除只要列有一個以上NaN
```

```python
dataFrame.dropna(thresh=1) #有一個Nan的刪除
```

```python
dataFrame.dropna(thresh=4, inplace=True) #刪除有4個NaN的
```

```python
dataFrame[-20:]
```

```python
#填滿NaN
dataFrame.fillna(0,inplace=True)
dataFrame
```
