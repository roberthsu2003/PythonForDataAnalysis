```python
#延續example3結果
#這cell是example3所做的功能

import numpy as np
import pandas as pd

dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame.rename(columns={'statistic_yyy':'年','site_id':'區域','people_total':'人口數','area':'土地面積','population_density':'人口密度'},inplace=True)
dataFrame.drop(0,inplace=True)
dataFrame.dropna(thresh=4,inplace=True)
dataFrame.fillna(0, inplace=True)

def transforming(x):
    if x == '… ':
        return 0
    else:
        return x
    
dataFrame = dataFrame.applymap(transforming)
dataFrame.index = dataFrame.index.map(lambda x: f'第{x}筆')
dataFrame
```

```python
#範例4現在開始
#搜尋資料
#先檢查資料類型
dataFrame.dtypes
```

```python
#全是字串
#astype()整欄轉型
dataFrame = dataFrame.astype({'人口數':np.int32,'土地面積':np.float32,'人口密度':np.float32})
dataFrame.dtypes
```

```python
#人口數大於300000
frame1=dataFrame[dataFrame['人口數'] >= 300000]
frame1
```

```python
#人口數大於300000,土地面績小於20
frame2 =frame1[frame1['土地面積'] < 20]
frame2               
```
