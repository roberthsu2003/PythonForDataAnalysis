```python
%matplotlib notebook
import matplotlib.pyplot as plt
days = range(0, 22, 3)
daysList = list(days)
celsius = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5, 32.6, 33.1]
plt.title('celsius/day')
plt.xlabel('days')
plt.ylabel('celsius')
plt.plot(daysList, celsius, 'g-')
plt.grid()
```
