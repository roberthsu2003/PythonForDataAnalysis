```python
#傳統函式
def f(x, y, z):
    return x + y + z

print('傳統函數處理')
f(20, 30 , 40)
f(x=20, z=40, y= 30)
```

```python
#匿名函式
fvar = lambda x, y, z: x + y + z
fvar(2, 30, 400)
```

```python
#傳統做法
def run_something_with_args(func, arg1, arg2):
    func(arg1, arg2)

def add_args(a1,b1):
    print(a1 + b1)

run_something_with_args(add_args, 10, 20)
```

```python
#lambda作法
run_something_with_args(lambda a1,b1:print(a1+b1), 10, 20)
```
