```python
#數學運算和資料對齊
import numpy as np
import pandas as pd

s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
s1
```

```python
s1 * 2
```

```python
# series 和 series運算
s1
```

```python
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
s2
```

```python
#運算時使用索引標籤對齊
s1 + s2
```

```python

#dataFrame數學運算時同時對齊row和columns標籤
df1 = pd.DataFrame(np.arange(9.).reshape((3, 3)),
                  columns=list('bcd'),
                  index=['台北','台中','台南'])
df2 = pd.DataFrame(np.arange(12.).reshape((4, 3)),
                  columns = list('bde'),
                  index=['高雄','台北','台中','台南'])
df1
```

```python
df2
```

```python
df1 + df2
```

```python
#相同row,不同的column標籤對齊
df1 = pd.DataFrame({'A':[1, 2]})
df2 = pd.DataFrame({'B':[3, 4]})
df1
```

```python
df2
```

```python
df1 + df2
```

```python
#數學運算實體方法
df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)),
                  columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape((4, 5)),
                  columns=list('abcde'))
df1
```

```python
df2
```

```python
df1 + df2
```

```python
#Function 和 Mapping
frame = pd.DataFrame(np.random.randn(4,3),
                    columns=list('bde'),
                    index=['台北','台中','台南','高雄'])
frame
```

```python
#numpy有提供一些數學function
np.abs(frame)
```

```python
#dataFrame實體方法apply(function)
frame
```

```python
def max_min(x):
    return x.max() - x.min()

frame.apply(max_min)
```

```python
frame.apply(lambda x: x.max() - x.min())
```

```python
frame.apply(lambda x:x.max() - x.min(),axis=1)
```
