```python
#一個一個讀取groupby物件內的值
import numpy as np
import pandas as pd

df = pd.DataFrame({'key1':['a', 'a', 'b', 'b', 'a'],
                  'key2':['one', 'two', 'one', 'two', 'one'],
                  'data1': np.random.randn(5),
                  'data2':np.random.randn(5)})
df
```

```python
grouped = df.groupby('key1')
for name, group in grouped:
    print(name)
    print(group)
```

```python
grouped = df.groupby(['key1', 'key2'])
for name, group in grouped:
    print(name)
    print(group)
    
```
