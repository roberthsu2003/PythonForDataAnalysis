```python
%matplotlib notebook
import matplotlib.pyplot as plt
import random
data = [random.randint(50,100) for _ in range(5)]
plt.figure(figsize=(8,5),num=3)
plt.plot(data)
plt.show()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
x = ['chinese', 'english', 'math', 'history', 'discover']
y1 = [1, 2, 5, 6, 7]
y2 = [4, 5, 6, 7, 8]
plt.plot(x,y1,x,y2)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.plot(['chinese', 'english', 'math', 'history', 'discover'],[1, 4, 9, 16, 25],'ro')
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.axis([-1,5,0,30])
plt.plot(['chinese', 'english', 'math', 'history', 'discover'],[1, 4, 9, 16, 25],'ro')
plt.grid()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
days = range(0,22,3)
celsius = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5, 32.6, 33.1]
plt.plot(days,celsius)
plt.xlabel("Day")
plt.ylabel('Celsius')
plt.title('Day-Celsius')
```
