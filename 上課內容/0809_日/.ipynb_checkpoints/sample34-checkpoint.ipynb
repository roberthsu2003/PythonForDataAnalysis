```python
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=10)
#font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
```

```python
import numpy as np
import pandas as pd

xlsx = pd.ExcelFile('總統.xls')
all = pd.read_excel(xlsx, '中　央')
all
```

```python
df=all.iloc[5:,0:4]
df.name = '第14任總統、副總統選舉候選人得票數一覽表'
df.columns = ['縣市','朱立倫','蔡英文','宋楚瑜']
df
```

```python
df=df.set_index('縣市')
df.columns.name = '總統候選人'
df
```

```python
df2=df.sum()
df2
```

```python
figure2, ax2= plt.subplots()
df2.plot(ax = ax2, kind='bar', color=['b','g','orange'])
ax2.set_xticklabels(['朱立倫','蔡英文','宋楚瑜'], fontproperties=font, rotation=30)
ax2.set_xlabel('總統候選人',fontproperties=font)
for p in ax2.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax2.annotate('{:}票'.format(height),(x,y+height+50000),fontproperties=font)
    
```

```python
#總得票率
df3 = df2 / df2.sum()
df3
```

```python
#顯示得票率
figure3, ax3 = plt.subplots()
df3.plot(ax = ax3, kind='barh', color=['b','g','orange'])
ax3.set_title('2016得票率',fontproperties=font)
ax3.set_ylabel('總統候選人',fontproperties=font)
ax3.set_yticklabels(df3.index,fontproperties=font)
for p in ax3.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax3.annotate('{:.2%}'.format(width), (width+0.01,y+height/2))
```
