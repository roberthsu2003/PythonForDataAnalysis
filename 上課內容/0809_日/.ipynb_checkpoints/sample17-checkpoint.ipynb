```python
%matplotlib notebook
import numpy as np
import math
import matplotlib.pyplot as plt


t = np.arange(0, 2.5, 0.1)
y1=np.sin(t * math.pi)
y2=np.sin(t * math.pi + math.pi/2)
y3=np.sin(t * math.pi - math.pi/2)
plt.plot(t,y1,'b*',t,y2,'g^',t,y3,'ys')
```

```python
#å±•ç¤ºç·šçš„æ¨£å¼
%matplotlib notebook
plt.plot(t, y1, 'b--', t, y2, 'g', t, y3, 'r-.')
```

```python
#subplot()å¯é¡¯ç¤º1å€‹figure å¤šå€‹Axes
#ç¬¬ä¸€å€‹åƒæ•¸åž‚ç›´åˆ†å‰²
#ç¬¬äºŒå€‹åƒæ•¸æ°´å¹³åˆ†å‰²
#ç¬¬ä¸‰å€‹é¸æ“‡é‚£ä¸€å€‹
%matplotlib notebook
t = np.arange(0, 5, 0.1)
y1 = np.sin(2 * np.pi * t)
y2 = np.sin(2*np.pi*t)

plt.subplot(211)
plt.plot(t, y1, 'b-.')
plt.subplot(212)
plt.plot(t, y2, 'r--')
```

```python
#subplot()å¯é¡¯ç¤º1å€‹figure å¤šå€‹Axes
#ç¬¬ä¸€å€‹åƒæ•¸åž‚ç›´åˆ†å‰²
#ç¬¬äºŒå€‹åƒæ•¸æ°´å¹³åˆ†å‰²
#ç¬¬ä¸‰å€‹é¸æ“‡é‚£ä¸€å€‹
%matplotlib notebook
t = np.arange(0, 1, 0.01)
y1 = np.sin(2 * np.pi * t)
y2 = np.sin(2*np.pi*t)

plt.subplot(121)
plt.plot(t, y1, 'b-.')
plt.subplot(122)
plt.plot(t, y2, 'r--')
```
