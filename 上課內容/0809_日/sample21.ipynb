```python
import csv
import pandas as pd

with open('新北市食品工廠清冊.csv','r',encoding='utf-8') as file:
    rows = csv.reader(file)
    dataList = list(rows)
    
    
titles = dataList.pop(0)
print(titles)
dataFrame = pd.DataFrame(dataList,columns=titles)
dataFrame.columns = ['工廠', '工廠登記編號','地址', '統編', '97x', '97y','緯度', '經度']
del dataFrame['97x']
del dataFrame['97y']
dataFrame
```

```python
# 請處理(新北市公共自行車租賃系統.json)
import json


with open('新北市公共自行車租賃系統.json') as file:
    data=json.load(file)
    
dataFrame=pd.DataFrame(data,columns=['sno', 'sarea', 'sna', 'ar', 'tot', 'sbi', 'bemp', 'lat','lng'])
dataFrame.columns = ['編號', '區域', '站名', '地址', '全部數量', '可借', '可還', '緯度', '經度']
dataFrame
```
