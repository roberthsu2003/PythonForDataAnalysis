## 第14任總統、副總統選舉候選人得票數一覽表

```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
```

```python
#讀取excel檔
xlsx = pd.ExcelFile('examples/總統.xls')

#取出一個sheet
all = pd.read_excel(xlsx,'中　央')
all
```

```python
#slice出所需要分析的資料
df = all.iloc[5:,:4]

#設定DataFrame的名字
df.name = '第14任總統、副總統選舉候選人得票數一覽表'

#重設欄位名稱
df.columns = ['縣市','朱立倫','蔡英文','宋楚瑜']

df
```

```python
#由於subset後,index number不對,使用reset_index()重新得到新的RangeIndex
df.reset_index().drop('index',axis=1)
```

```python
#轉換 '縣市' 欄位成為index
df=df.set_index('縣市')
df.columns.name = '總統候選人'
df
```

## 總得票數

```python
#使用sum()取得加總票數(series)
df2 = df.sum()
df2
```

```python
#建立bar圖形
figure2, ax2 = plt.subplots()
df2.plot(ax = ax2,kind='bar',color=['b','g','orange'])
ax2.set_xticklabels(df2.index,fontproperties=font,rotation=30)
ax2.set_xlabel('總統候選人',fontproperties=font)
for p in ax2.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax2.annotate('{:}票'.format(height), (x, y+ height + 100000),fontproperties=font)
```

## 總得票率

```python
#使用Series 數學運算,得到得票率
df3 = df2 / df2.sum()
df3
```

```python
#顯示得票率

figure3, ax3 = plt.subplots()
df3.plot(ax=ax3,kind='barh',color=['b','g','orange'])
ax3.set_title('2016得票率',fontproperties=font)
ax3.set_ylabel('總統候選人',fontproperties=font)
ax3.set_yticklabels(df3.index,fontproperties=font)
for p in ax3.patches:
    width, height = p.get_width(), p.get_height()
    print(width)
    x, y = p.get_xy()
    ax3.annotate('{:.0%}'.format(width), (width+0.01, y+height/2))
```

## 各縣市得票率

```python
#使用dataFrame的apply()取得新的得票率

def aa(arr):
    total = arr['朱立倫'] + arr['蔡英文'] + arr['宋楚瑜']
    arr['朱立倫'] = arr['朱立倫'] / total
    arr['蔡英文'] = arr['蔡英文'] / total
    arr['宋楚瑜'] = arr['宋楚瑜'] / total    
    return arr

df1 = df.apply(aa,axis=1)
df1
```

```python
#顯示各縣市得票率
figure1, ax1 = plt.subplots()
df1.plot(ax = ax1,kind='bar',color=['b', 'g', 'orange'])
ax1.set_xticklabels(df1.index,fontproperties=font)
ax1.legend(prop=font)
ax1.set_title('各縣市得票率',fontproperties=font)
```

```python
#取得蔡英文win and lose
def win(x):
    if x['蔡英文'] > x['朱立倫']:
        return '蔡WIN'
    else:
        return '蔡LOSE'
df5 = df1.drop(columns='宋楚瑜') 
df5['結果'] = df5.apply(win,axis=1)
df5.sort_values(by='結果',ascending=False,inplace=True)
df5
```

```python
df5.reset_index(inplace=True)
df5.set_index(['結果','縣市'])
```
