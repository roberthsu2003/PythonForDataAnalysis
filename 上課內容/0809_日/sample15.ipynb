```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

days = np.array(['1/3','2/3','3/3','4/3','5/3','6/3','7/3','8/3'])
celsius = np.array([25.6, 23.2, 18.5, 28.3, 26.5, 30.5, 32.6, 33.1])
plt.plot(days,celsius)
plt.show()
```
