```python
import numpy as np
import pandas as pd
dataFrame=pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame.iloc[:100]
```

```python
dataFrame.columns = dataFrame.iloc[0]
dataFrame
```

```python
dataFrame = dataFrame.iloc[1:-5]
dataFrame
```

```python
def a1(x):
    if len(x) > 4:
        return x[:3]
    else:
        return x
    
states=dataFrame['區域別'].map(lambda x:x[:3] if len(x)> 4 else x)
dataFrame['縣市'] = states
dataFrame
```

```python
dataFrame=dataFrame.reindex(columns=['縣市','年底人口數','土地面積'])
dataFrame
```

```python
dataFrame.dtypes
```

```python
def toInt(x):
    try:
        return int(x)
    except:
        return 0


dataFrame['年底人口數']=dataFrame['年底人口數'].map(toInt)
dataFrame
```

```python
dataFrame['土地面積']=dataFrame['土地面積'].map(lambda x:float(x))
dataFrame.
```
