```python
import pandas as pd
import numpy as np

frame = pd.DataFrame(np.random.randn(4,3),
                    columns=list('bde'),
                    index=['台北','台中','台南','高雄'])
frame
```

```python
def fun1(s):    
    return s

frame.apply(fun1)
```

```python
def fun2(s):    
    return s.max()

frame.apply(fun2)
```

```python
def fun3(s):
    return pd.Series([s.max(), s.min()], index=['最大值','最小值'])

frame.apply(fun3)
```

```python
#練習建立學生成績單
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
scores_pd = pd.DataFrame(scores,columns=['國文', '英文', '數學', '地理', '歷史'],index=range(1,51))
scores_pd.index.name="學號"
scores_pd.columns.name="科目"
scores_pd
```

```python
#使用apply(),做統計動作,產生新的DataFrame

def cal1(s):
    return pd.Series([s.min(),s.max(),s.median()],index=['最低分','最高分','中間值'])

scores_pd.apply(cal1)
```

```python
#使用lambda代替function
scores_pd.apply(lambda s: pd.Series([s.min(),s.max(),s.median()],index=['最低分','最高分','中間值']))
```
