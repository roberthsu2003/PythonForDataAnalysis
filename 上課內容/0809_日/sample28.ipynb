```python
#AQI
#下載檔案
#https
import urllib3
import certifi
CSV_URL = "http://opendata.epa.gov.tw/webapi/Data/REWIQA/?$orderby=SiteName&$skip=0&$top=1000&format=csv"
http = urllib3.PoolManager()
response=http.request('GET',CSV_URL)
if response.status == 200:
    print("下載成功")
    with open('空氣品質指標.csv','wb') as file:
        file.write(response.data)
        file.close()
else:
    print("下載失敗")
```

```python
import pandas as pd 
aqiDataFrame = pd.read_csv('空氣品質指標.csv')
aqiDataFrame.dtypes
```

```python
aqi=aqiDataFrame.reindex(columns=['SiteName','AQI','PM2.5_AVG','PublishTime'])
aqi
```

```python
min = aqi['AQI'].min()
max = aqi['AQI'].max()
mean = aqi['AQI'].mean()
```

```python
#最好
aqi[aqi['AQI'] == min]
```

```python
#最不好
aqi[aqi['AQI'] == max]
```

```python
#比平均差的
aqi[aqi['AQI'] >= mean]
```

```python
#比平均好的
aqi[aqi['AQI'] < mean]
```
