```python
scores = []
sum = 0
nums = 5
for i in range(nums):
    i += 1
    print("請輸入第{}位學生的資料".format(i))
    scores.append(int(input()))

for score in scores:
    sum += score
ave = sum / nums
print("全班總成績為:{0:d},平均為{1:.2f}".format(sum, ave))
```
