```python
import numpy as np
import pandas as pd

scores = np.random.randint(50, 101, size=(50, 5))
scores_df = pd.DataFrame(scores,
                         columns=['國文','英文','數學','地理','歷史'],
                        index = range(1, 51))
scores_df.index.name='座號'
scores_df.columns.name='科目'
scores_df
```

```python
scores_df.head()
```

```python
scores_df.sum()
```

```python
scores_df.mean()
```

```python
for index in range(1,51):
    print('學號{:d},總分為{:d},平均為:{:.2f}'.format(index,scores_df.loc[index].sum(),scores_df.loc[index].mean()))
```
