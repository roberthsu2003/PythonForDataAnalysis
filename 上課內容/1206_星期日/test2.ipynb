```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
index = [('台北', 2019), ('台北', 2020),
         ('台中', 2019), ('台中', 2020),
         ('台南', 2019), ('台南', 2020)]
populations = [33871648, 37253956,
               18976457, 19378102,
               20851820, 25145561]
index = pd.MultiIndex.from_tuples(index, names=['縣市','年'])
pop = pd.Series(populations,index=index)
pop
```

```python
pop_df = pop.unstack()
pop_df
```

```python
#import FontProperties
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#window font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)

figure, ax1 = plt.subplots(1,1)
pop_df.plot(ax=ax1,kind='bar',color=['r', 'g'])
ax1.set_xticklabels(pop_df.index,fontproperties=font)
ax1.legend(prop=font)
ax1.set_title('2019,2020城市人口',fontproperties=font)
ax1.set_xlabel('縣市',fontproperties=font)
ax1.set_yticklabels(['0','5佰萬','1仟萬','1仟5佰萬','2仟萬','2仟5佰萬','3仟萬','3仟5佰萬'],fontproperties=font)
for p in ax1.patches:
    width, height = p.get_width(), p.get_height()
    print('width',width)
    print('height',height)
    x, y = p.get_xy()
    print('x',x)
    print('y',y)
    print('x+width/2',x+width/2)
    print('y+height', x+height)
    ax1.annotate('{:}'.format(height), (x+width/2-width/2, y+height+height/100), fontproperties=font)
```
