```python
import numpy as np
import pandas as pd

#美國球賽門票收入
tips = pd.read_csv('tips.csv')
tips
```

```python
tips.columns = ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
tips.drop('大小',axis=1,inplace=True)
```

```python
tips
```

```python
tips.apply(lambda x:x['小費'] / x['總票價'],axis=1)
```

```python
tips['小費'] / tips['總票價']
```

```python
tips['小費比列'] = tips['小費'] / tips['總票價']
tips
```

```python
grouped = tips.groupby(['日期','吸煙者'])
grouped
```

```python
grouped_series=grouped['小費']
```

```python
grouped_series.agg('mean')
```

```python
grouped_series.agg(['mean'])
```

```python
grouped_series.agg(['mean','sum','std'])
```

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()

grouped_series.agg(['mean','std',peak_to_peak])
```
