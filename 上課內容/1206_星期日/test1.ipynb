```python
import numpy as np
import pandas as pd
df = pd.DataFrame({'key1':['a', 'a', 'b', 'b', 'a'],
                  'key2':['one', 'two', 'one', 'two', 'one'],
                  'data1': np.random.randn(5),
                  'data2':np.random.randn(5)})
df
```

```python
grouped = df['data1'].groupby(df['key1'])
grouped
```

```python
grouped.mean()
```

```python
means=df['data1'].groupby([df['key1'],df['key2']]).mean()
means
```

```python
means.unstack()
```

```python
states = np.array(['台北','台中','台中','台北','台北'])
years = np.array([2005, 2005, 2006, 2005, 2006])
df['data1'].groupby([states, years]).mean()
```

```python
import numpy as np
import pandas as pd

tips = pd.read_csv('tips.csv')
tips
```

```python
tips.dtypes
```

```python
tips.shape
```

```python
tips.columns =['總票價', '小費', '吸煙者', '日期', '時間', '大小']
tips
```

```python
tips['小費比例'] = tips['小費'] / tips['總票價']
tips
```

```python
grouped = tips.groupby(['日期','吸煙者'])
grouped
```

```python
grouped_pct = grouped['小費']
```

```python
grouped_pct.agg('mean')
```

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()

grouped_pct.agg(['mean','std',peak_to_peak])
```

```python
grouped_pct.agg([('平均','mean'),('標準差',np.std)])
```

```python
result = grouped[['小費','總票價']].agg([('平均','mean'),('數量','count'),('最大值','max')])
result
```
