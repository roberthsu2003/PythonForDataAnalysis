```python
#list轉為ndarray
a = np.array([1, 3, 5, 7, 9])
b = np.array([3, 5, 6, 7, 9])
c = a + b
print(c)

print(type(c))
```

```python
[1, 3, 5, 7, 9] + [3, 5, 6, 7, 9]
```

```python
print(c)
print(c.ndim)
print(c.shape)
print(c.size)
print(c.dtype)
```
