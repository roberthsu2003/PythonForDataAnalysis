```python
import numpy as np
import pandas as pd
scores = np.random.randint(low=50,high=101,size=(50,5))
scores_pd = pd.DataFrame(scores,
                         index=range(1,51),
                         columns = ['國文', '英文', '數學', '地理', '歷史']
                        )
scores_pd
```

```python
sum_values = scores_pd.sum(axis=1)
mean_values = scores_pd.mean(axis=1)
rank_values = sum_values.rank(method='first',ascending=False)
rank_values
```

```python
scores_pd['總分'] = sum_values
scores_pd['平均'] = mean_values
scores_pd['排名'] = rank_values
```

```python
scores_pd
```

```python
scores_pd.sort_values(by='排名')
```
