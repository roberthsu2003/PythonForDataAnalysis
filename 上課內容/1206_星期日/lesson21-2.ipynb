```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
```

```python
xlsx = pd.ExcelFile('總統.xls')
all_dataFrame = pd.read_excel(xlsx,'中　央')
all_dataFrame.dtypes
```

```python
all_dataFrame.info()
```

```python
all_dataFrame
```

```python
df = all_dataFrame.iloc[5:,:4]
df.name = '第14任總統、副總統選舉候選人得票數一覽表'
df.columns = ['縣市','朱立倫','蔡英文','宋楚瑜']
df
```

```python
df.reset_index(drop=True,inplace=True)
df
```

```python
df.set_index('縣市',inplace=True)
df
```

```python
df.columns.name = '總統候選人'
df
```

```python
seObj = df.sum()
seObj
```

```python
#建立bar圖形
figure2, ax2 = plt.subplots()
seObj.plot(ax = ax2, kind='bar', color=['b','g','orange'])
ax2.set_xticklabels(seObj.index,fontproperties=font,rotation=30)
ax2.set_xlabel('總統候選人',fontproperties=font)
for p in ax2.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax2.annotate('{:}票'.format(height), (x+width/5, y+ height + 100000),fontproperties=font)
```

```python
seObj
seObj.sum()
```

```python
seObjPercent = seObj / seObj.sum()
seObjPercent
```

```python
figure3, ax3 = plt.subplots()
seObjPercent.plot(ax=ax3, kind='barh',color=['b', 'g', 'orange'])
ax3.set_title('2016得票率',fontproperties=font)
ax3.set_ylabel('總統候選人',fontproperties=font)
ax3.set_yticklabels(seObjPercent.index,fontproperties=font)
for p in ax3.patches:
    width, height = p.get_width(), p.get_height()
    print(width)
    x, y = p.get_xy()
    ax3.annotate('{:.0%}'.format(width), (width+0.01, y+height/2))
```
