```python
#建立bar圖形
figure2, ax2 = plt.subplots()
seObj.plot(ax = ax2, kind='bar', color=['b','g','orange'])
ax2.set_xticklabels(seObj.index,fontproperties=font,rotation=30)
ax2.set_xlabel('總統候選人',fontproperties=font)
for p in ax2.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax2.annotate('{:}票'.format(height), (x+width/5, y+ height + 100000),fontproperties=font)
```

```python
def getPercent(ser): 
    total = ser.sum()
    ser['朱立倫'] = ser['朱立倫'] / total
    ser['蔡英文'] = ser['蔡英文'] / total
    ser['宋楚瑜'] = ser['宋楚瑜'] / total
    return ser

def getPercent1(ser):
    total = ser.sum()
    se = pd.Series([ser['朱立倫']/total,
                    ser['蔡英文']/total,
                    ser['宋楚瑜']/total,
                    total],   
                   index=['朱立倫','蔡英文','宋楚瑜','縣市總票數'])
    return se

def getPercent2(ser):
    total = ser.sum()
    se = pd.Series([ser['蔡英文']/total,
                    ser['朱立倫']/total,
                    ser['宋楚瑜']/total],                       
                   index=['蔡英文','朱立倫','宋楚瑜'])
    return se

df1 = df.apply(getPercent2,axis=1)
```

```python
figure4, ax4 = plt.subplots()
df1.plot(ax = ax4,kind='bar',color=['g', 'b', 'orange'])
ax4.set_xticklabels(df1.index,fontproperties=font)
ax4.legend(prop=font)
ax4.set_title('各縣市得票率',fontproperties=font)
```

```python
df1
```

```python
def getWin(ser):
    if ser['蔡英文'] > ser['朱立倫']:
        return '蔡win'
    else:
        return '朱win'
    
df2 = df1.drop(columns='宋楚瑜')

df2['結果'] = df2.apply(getWin,axis=1)
df2.sort_values(by='結果',inplace=True,ascending=False)
df2
```
