```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(start=-3,stop=3,num=30)
y1 = 2*x + 1
y2 = x ** 2
plt.plot(x, y2)
plt.plot(x, y1, color="red", linewidth=1.0, linestyle="--")
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel("i am x")
plt.ylabel("i am y")
```
