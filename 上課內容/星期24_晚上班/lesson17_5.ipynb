```python
import pandas_datareader.data as web
#台積電(2330.TW),聯電(2303.TW),聯發科(2454.TW),鴻海(2317.TW)
web.get_data_yahoo('2330.TW')
```

```python
all_data = {ticker:web.get_data_yahoo(ticker) for ticker in ['2330.TW','2303.TW','2454.TW','2317.TW']}
all_data
```

```python
all_data1= {ticker:data['Adj Close'] for ticker, data in all_data.items()}
all_data1
    
```

```python
import pandas as pd
price = pd.DataFrame(all_data1)
price.columns = ['台積電','聯電','聯發科','鴻海']
price
```

```python
pctPrice = price.pct_change()
pctPrice
```

```python
pctPrice['台積電'].corr(pctPrice['鴻海'])
```
