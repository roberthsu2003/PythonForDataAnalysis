```python
#練習建立學生成績單
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
scores_pd = pd.DataFrame(scores,columns=['國文', '英文', '數學', '地理', '歷史'],index=range(1,51))
scores_pd.index.name="學號"
scores_pd.columns.name="科目"
scores_pd
```

```python
scores_pd.sum()
```

```python
scores_pd.sum(axis=1)
```

```python
def myfun1(s):
    return pd.Series([s.max(),s.min(),s.median()],index=["最高分","最低分","中間分數"])

scores_pd.apply(myfun1)
```
