```python
import matplotlib.pyplot as plt
import numpy as np

figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
x_linespace = np.linspace(-3,3,50)
y1 = 2 * x_linespace + 1 
axes.plot(x_linespace,y1,'r--')

y2 = x_linespace ** 2
axes.plot(x_linespace,y2,'g--')

plt.show()
```

```python
range_0_5 = np.arange(0,5,0.1)
y1 = np.sin(2 * np.pi * range_0_5)
y2 = np.cos(2 * np.pi * range_0_5)

figure = plt.figure(figsize=(8,15))
ax1 = figure.add_subplot(2,1,1)
ax1.plot(range_0_5,y1,'b-.')
ax2 = figure.add_subplot(2,1,2)
ax2.plot(range_0_5,y2,'r-.')

plt.show()
```

```python
index = ['apple', 'banana', 'orange', 'tomato', 'guava']
values = np.array([5, 7, 3, 4, 6]) * 1024

fig = plt.figure(figsize=(8, 5))
axes = fig.add_subplot(1,1,1)
axes.bar(index, values)
plt.show()
```

```python
index = np.arange(5)
values = np.array([5, 7, 3, 4, 6]) * 1024

fig = plt.figure(figsize=(8, 5))
axes = fig.add_subplot(1,1,1)
axes.bar(index, values)
axes.axis([-1,5,0,8000])
axes.set_xticks(index,['apple', 'banana', 'orange', 'tomato', 'guava'])
for i in index:
    axes.text(i-0.2,values[i]+200,values[i])
plt.show()
```
