```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8.0,5.0))
ax = fig.add_subplot(1,1,1)
ax.plot([0, 1, 2],[3, 4, 5])
plt.show()

```

```python
fig = plt.figure(figsize=(8, 5))
ax_1 = fig.add_subplot(1,3,1)
ax_2 = fig.add_subplot(1,3,2)
ax_3 = fig.add_subplot(1,3,3)

ax_1.set_title("title1")
ax_2.set_title("title2")
ax_3.set_title("title3")

plt.show()
```

```python
fig_1 = plt.figure(figsize=(8,5))
ax_1 = fig_1.add_subplot(1,1,1)
ax_1.plot([0,1,2],[3,4,1])

fig_2 = plt.figure(figsize=(5,8))
ax_2 = fig_2.add_subplot(1,1,1)
ax_2.plot([2, 3, 6],[2,1,2])
plt.show()
```

```python
data = [-1, -4.3, 15, 21, 31]
xticks = ['a', 'b', 'c', 'd', 'e']
fig = plt.figure(figsize=(8,5))
ax_1 = fig.add_subplot(1,1,1)
ax_1.plot(xticks,data)
plt.show()
```

```python
student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2 = {'chinese':92, 'english':60, 'math':85, 'history':62, 'discover':73}
subjects1 = list(student1.keys())
scores1= list(student1.values())
subjects2 = list(student2.keys())
scores2 = list(student2.values())
print(subjects1)
print(scores1)

fig = plt.figure(figsize=(8,5),dpi=250,facecolor='#cccccc')
ax = fig.add_subplot(1,1,1)
ax.plot(subjects1,scores1,subjects2,scores2)
plt.show()
```
