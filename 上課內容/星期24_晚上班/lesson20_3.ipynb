```python
import numpy as np
import pandas as pd

#美國球門票收入
tips = pd.read_csv('tips.csv')
tips
```

```python
tips.columns = ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
tips
```

```python
tips.info()
```

```python
tips
```

```python
#增加小費比例
tips['小費比例'] = tips['小費'] / tips['總票價']
tips
```

```python
#取得小費比例最高的前6筆
tips.sort_values(by='小費比例',ascending=False).iloc[:6]
```

```python
#建立自訂function,小費比例最高的前6筆
def top(df,n=6,columns='小費比例'):
    return df.sort_values(by=columns,ascending=False).iloc[:n]

top(tips)
```

```python
#請取出吸煙者和非煙者,小費比例最高的前6筆
tips.groupby(by='吸煙者').apply(top,n=3)
```

```python
tips.groupby(by=['吸煙者','日期']).apply(top,n=1)
```
