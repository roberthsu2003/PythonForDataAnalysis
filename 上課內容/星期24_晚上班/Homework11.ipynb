```python
#歷年國內主要觀光遊憩據點遊客人數月別統計.csv
import numpy as np
import pandas as pd

dataFrame = pd.read_csv('歷年國內主要觀光遊憩據點遊客人數月別統計.csv')
dataFrame1=dataFrame.drop(columns=['類型','細分'])
dataFrame1
```

```python
dataFrame2=dataFrame1.set_index(['年別','縣市別'])
dataFrame2
```

```python
dataFrame2.groupby(by=['縣市別']).sum(True)
```

```python
dataFrame2.groupby(by=['年別','縣市別']).sum(True)
```

```python
dataFrame3 = dataFrame1.set_index(['年別'])
dataFrame3
```

```python
dataFrame3.groupby(by='年別').sum(numeric_only=True)
```
