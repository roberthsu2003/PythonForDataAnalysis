```python
import numpy as np
import pandas as pd

dataFrame = pd.read_csv('world.csv')
dataFrame.iloc[:10]
```

```python
allDataFrame=dataFrame.reindex(columns=['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數'])
allDataFrame
```

```python
allDataFrame.info()
```

```python
#搜尋[]國家==台灣
taiwanFrame = allDataFrame[allDataFrame['國家'] == '台灣']
taiwanFrame
```

```python
#台灣2022-01-01~2022-05-31
taiwan2022=taiwanFrame[(taiwanFrame['日期'] >= '2022-01-01') & (taiwanFrame['日期'] <= '2022-05-31')]
taiwan2022
```

```python
#搜尋[]國家==日本
japanFrame = allDataFrame[allDataFrame['國家'] == '日本']
japanFrame
```

```python
japan2022=japanFrame[(japanFrame['日期'] >= '2022-01-01') & (japanFrame['日期'] <= '2022-05-31')]
japan2022
```

```python
taiwan=taiwan2022[['新增確診數']].rename(columns={'新增確診數':'台灣2022-01-05'})
taiwan
```

```python
japan=japan2022[['新增確診數']].rename(columns={'新增確診數':'日本2022-01-05'})
japan
```
