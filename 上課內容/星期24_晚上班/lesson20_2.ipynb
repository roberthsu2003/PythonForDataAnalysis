```python
import numpy as np
import pandas as pd

dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame
```

```python
#如果欄位是整數,但傳出的是object,整欄無法全部轉成整數
dataFrame.info()
```

```python
#改變欄位名
dataFrame.columns = dataFrame.iloc[0]
dataFrame
```

```python
#清除第一列
dataFrame1 = dataFrame.drop(labels=0)
dataFrame1
```

```python
dataFrame1.iloc[-10:]
```

```python
#移除後面不要的列
dataFrame2 = dataFrame1.dropna()
dataFrame3=dataFrame2.drop(labels=[369,370])
dataFrame3
```

```python
dataFrame3.info()
```

```python
dataFrame4 = dataFrame3.astype(dtype={'年底人口數':np.int32,'土地面積':np.float32,'人口密度':np.int32})
dataFrame4
```

```python
dataFrame4.info()
```

```python
#[]搜尋或選取
#搜尋新北市三重區
dataFrame4[dataFrame4['區域別'] == '新北市三重區']
```

```python
#如果整欄是字串,可以使用str
dataFrame4[dataFrame4['區域別'].str.contains('臺北市')]
```

```python
#取出縣市
dataFrame4['縣市'] = dataFrame4['區域別'].str[:3]
dataFrame4
```

```python
dataFrame5 = dataFrame4.groupby('縣市')[['年底人口數','土地面積']].sum()
dataFrame5
```

```python
#重新手動排序index
dataFrame5.reindex(['臺北市','新北市'])
```
