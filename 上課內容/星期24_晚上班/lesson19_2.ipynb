```python
import numpy as np
import pandas as pd

dataFrame = pd.read_csv('歷年國內主要觀光遊憩據點遊客人數月別統計.csv')
dataFrame
```

```python
dataFrame.info()
```

```python
dataFrame.isna().sum()
```

```python
dataFrame.fillna(0,inplace=True)
```

```python
dataFrame.info()
```

```python
dataFrame1 = dataFrame.reindex(columns=['年別','類型','觀光遊憩區','縣市別','合計'])
dataFrame1
```

```python
dataFrame2=dataFrame1.set_index(['年別','類型','觀光遊憩區','縣市別'])
dataFrame3=dataFrame2.stack()
dataFrame4=dataFrame3.groupby(level=['年別','縣市別']).sum()
dataFrame4.unstack(level='年別')
```

```python
dataFrame5 = dataFrame3.groupby(level=['年別','類型','縣市別']).sum()
dataFrame6=dataFrame5.unstack(level='類型',fill_value=0)
dataFrame6
```
