```python
import numpy as np
import pandas as pd

dataFrame = pd.read_csv('world.csv')
dataFrame
```

```python
grouped = dataFrame.groupby(by='洲名')
grouped['新增死亡數'].agg([('總死亡數','sum')])
```

```python
grouped = dataFrame.groupby(by=['洲名','國家'])
grouped['新增死亡數'].agg([('總死亡數','sum'),('最高死亡數','max')])
```

```python
grouped = dataFrame.groupby(by=['洲名','國家'])
grouped['新增死亡數'].agg([('總死亡數','sum'),('最高死亡數','max')]).loc['亞洲']
```
