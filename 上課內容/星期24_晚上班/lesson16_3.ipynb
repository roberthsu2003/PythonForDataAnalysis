```python
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
scores
```

```python
scores_frame = pd.DataFrame(scores,
                            columns=['國文', '英文', '數學', '地理', '歷史'],
                            index=range(1,51))
scores_frame.index.name="學號"
scores_frame.columns.name="科目"
scores_frame
```

```python
sum_s=scores_frame.sum(axis=1)
average_s = scores_frame.mean(axis=1)
rank_s = sum_s.rank(ascending=False,method='first')
rank_s
```

```python
scores_frame['總分'] = sum_s
scores_frame['平均'] = average_s
scores_frame['排名'] = rank_s
scores_frame
```
