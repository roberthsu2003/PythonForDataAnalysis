```python
def student(nums):    
    scores = []
    sum = 0

    for i in range(nums):
        inputValue = int(input(f"請輸入第{i+1}位學生的資料:"))
        scores.append(inputValue)

    for score in scores:
        sum += score

    ave = sum / nums
    print("全班總成績為{0:d}分,平均為{1:.2f}分".format(sum,ave))

student(3)
```
