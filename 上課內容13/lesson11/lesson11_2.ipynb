```python
import numpy as np
import pandas as pd


data=np.random.randint(50,101,size=(50,5))
students = pd.DataFrame(data,columns=['國文','英文','數學','地理','歷史'],index=range(1,51))
students
```

```python
student_sum = students.sum(axis=1)
student_average = students.mean(axis=1)
students['總分'] = student_sum
students['平均'] = student_average
students
```

```python
students.to_csv('students.csv')
```
