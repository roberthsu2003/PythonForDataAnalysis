```python
import numpy as np
import pandas as pd
result = pd.read_csv('各鄉鎮市區人口密度-2.csv')
result.columns = result.loc[0]
result1 = result.drop(0)
result1
```

```python
result2=result1.dropna()
result2
```

```python
result3 = result2.drop([369,370])
result3
```

```python
result3.dtypes
```

```python
result4=result3.astype({'統計年':int,'年底人口數':int,'土地面積':float,'人口密度':int})
result4.dtypes
```

```python
def getState(x):
    return x[:3]

result4['縣市'] = result4['區域別'].map(getState)
result4
```

```python
result4['區域別'] = result4['區域別'].map(lambda x:x[3:])
result4
```

```python
result5=result4.set_index(['縣市','區域別'])
result5
```

```python
result6=result5.reindex(columns=['年底人口數','土地面積','人口密度'])
result6
```

```python
result7 = result6.groupby(level='縣市').sum()
result7
```

```python
import matplotlib.pyplt as plt
figure = plt.figure()
axes=figure.addsubplot(1,1,1)
result7['年底人口數'].plot()
```
