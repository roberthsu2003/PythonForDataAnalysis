```python
import matplotlib.pyplot as plt
```

```python
data = [-1, -4.3, 15, 21, 31]
plt.figure(num=3,figsize=(8,5))
plt.plot(data)
plt.show()
```

```python
x = ['chinese', 'english', 'math', 'history', 'discover']
y1 = [97, 68, 75, 62, 99]
y2 = [99, 78, 82, 74, 90]

plt.figure(4,figsize=(8,5))
plt.plot(x,y1,x,y2)
plt.show()
```

```python
x = ['chinese', 'english', 'math', 'history', 'discover']
y1 = [97, 68, 75, 62, 99]
y2 = [99, 78, 82, 74, 90]

plt.figure(5,figsize=(8,5))
plt.plot(x,y1,'ro')
plt.show()
```

```python
x = ['chinese', 'english', 'math', 'history', 'discover']
y1 = [97, 68, 75, 62, 99]
y2 = [99, 78, 82, 74, 90]

plt.figure(6,figsize=(8,5))
plt.plot(x,y1,'r-o', x, y2, 'g--')
plt.grid()
plt.show()
```

```python
plt.figure(7,figsize=(8,5))
plt.title('My First Plot')
plt.axis([0,5,0,20])
plt.plot([1, 2, 3, 4],[1,4,9,16],'ro')
plt.show()
```

```python
plt.figure(8,figsize=(8,5))
days = range(0, 22, 3)
celsius = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5, 32.6, 33.1]
plt.title('My First Plot')
plt.xlabel('Day')
plt.ylabel('Celsius')
plt.plot(days,celsius,'ro')
plt.show()
```

```python
plt.figure(9,figsize=(8,5))
index = ['apple', 'banana', 'orange', 'tomato', 'guava']
values = [5, 7, 3, 4, 6]
plt.bar(index,values)
plt.show()
```
