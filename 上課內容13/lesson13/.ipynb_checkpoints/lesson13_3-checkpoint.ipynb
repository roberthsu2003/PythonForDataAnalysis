```python
import pandas as pd
import numpy as np

dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame

#使用dtypes屬性查詢dataFrame所有的欄的資料類型
dataFrame.dtypes

#使用shape屬性查詢欄列數量
dataFrame.shape

#改變欄位索引標籤名稱
dataFrame.columns = dataFrame.loc[0]
dataFrame

#查詢列索引
dataFrame.index

#移除第一列
dataFrame = dataFrame.drop(0)
dataFrame

#查看最後10行
dataFrame.iloc[-10:]

#使用drop刪除最後幾行
dataFrame.drop([371, 372, 373, 374, 375],inplace=True)
dataFrame

#建立新的索引列
#先用shape取出列的數量
rowCount, columnCount = dataFrame.shape
rowCount

#建立新的索引列
dataFrame.index = pd.Index(range(rowCount))
dataFrame

#使用reindex取出想要的欄位
dataFrame1 = dataFrame.reindex(columns=['區域別','年底人口數','土地面積'])
dataFrame1

#使用rename()實體方法改變欄位名稱
dataFrame1 = dataFrame1.rename(columns={'年底人口數':'人口數'})
dataFrame1
```

```python
dataFrame1.dtypes
```

```python
def format1(x):
    try:
        value = int(x)
    except:
        value = 0
    
    return value

newPopulation = dataFrame1['人口數'].map(format1)
dataFrame1['人口數'] = newPopulation
dataFrame1.dtypes
```

```python
def format2(x):
    return float(x)

newArea = dataFrame1['土地面積'].map(format2)
dataFrame1['土地面積'] = newArea
dataFrame1.dtypes
```

```python
dataFrame1
```

```python
dataFrame1[['人口數','土地面積']].apply(lambda x:pd.Series([x.max(),x.min()],index=['最大','最小']))
```
