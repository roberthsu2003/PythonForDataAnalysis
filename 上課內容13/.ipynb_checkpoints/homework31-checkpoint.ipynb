```python
import numpy as np
import pandas as pd

dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame
```

```python
dataFrame.columns = dataFrame.iloc[0]
dataFrame = dataFrame.drop(0)
```

```python
dataFrame[-20:]
```

```python
#刪除最後5筆
dataFrame = dataFrame[:-5]
dataFrame
```

```python
dataFrame['區域別'][-20:]
```

```python
states = dataFrame['區域別'].map(lambda x: x[:3] if len(x) > 4 else x)
dataFrame['縣市'] = states
dataFrame = dataFrame.reindex(columns=['縣市','年底人口數','土地面積'])
```

```python
dataFrame.info()
```

```python
def toInt(x):
    try:
        return np.int64(x)
    except:
        return 0

#dataFrame['年底人口數'].map(lambda x: int(x))
dataFrame['年底人口數'] = dataFrame['年底人口數'].map(toInt)
dataFrame['土地面積'] = dataFrame['土地面積'].map(lambda x: np.float64(x))
dataFrame
```

```python
dataFrame = dataFrame.reset_index(drop=True)
```

```python
grouped = dataFrame.groupby('縣市')
dict(list(grouped))['臺北市']
```

```python
result = grouped.sum()
result['人口密度'] = result['年底人口數'] / result['土地面積']
result.sort_values('年底人口數',ascending=False)
```
