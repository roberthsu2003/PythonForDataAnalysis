```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {'年廣告費投入':[12.5, 15.3, 23.2, 26.4, 33.5, 34.4, 39.4, 45.2, 55.4, 60.9],
            '月均銷售額':[21.2, 23.9, 32.9, 34.1, 42.5, 43.2, 49.0, 52.8, 59.4, 63.5]}
dataFrame = pd.DataFrame(data,index=[2020,2019,2018,2017,2016,2015,2014,2013,2012,2011])
plt.plot(dataFrame.values,'ko--')
```

```python
dataFrame['年廣告費投入'].corr(dataFrame['月均銷售額'])
```

```python
dataFrame['年廣告費投入'].cov(dataFrame['月均銷售額'])
```

```python
dataFrame.corr()
```

```python
import pandas_datareader.data as web
stocks = ['AAPL', 'IBM', 'MSFT', 'GOOG']
all_data = {ticker:web.get_data_yahoo(ticker) for ticker in stocks}
prices_python = {key:data['Adj Close'] for key,data in all_data.items()}
priceDataFrame = pd.DataFrame(prices_python)
priceDataFrame
returns=priceDataFrame.pct_change() 
```
