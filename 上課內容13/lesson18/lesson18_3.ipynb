```python
# 請處理(新北市公共自行車租賃系統.json)
# google搜尋online json viewer
#api接口:https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/json?page=0&size=1000

import requests
import numpy as np
import pandas as pd

url = 'https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/json?page=0&size=1000'
response = requests.get(url)
if response.status_code == 200:
    python_json = response.json()    

df = pd.DataFrame(python_json,columns=['sna','tot','sbi','sarea','ar','bemp'])

df=df.rename(columns={'sna':'站名',
                      'sarea':'區域',
                      'ar':'地址',
                      'tot':'數量',
                      'sbi':'可借',
                      'bemp':'可還'})
              
                   
```

```python
df
```

```python
df=df.astype({'數量':int,'可借':int,'可還':int})
df.dtypes
```

```python
df['故障'] = df['數量'] - (df['可借']+df['可還'])
df['故障率'] = df['故障'] / df['數量']
df
```
