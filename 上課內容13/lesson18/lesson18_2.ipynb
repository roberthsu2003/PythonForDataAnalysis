```python
import numpy as np
import pandas as pd

tips = pd.read_csv('tips.csv')
tips
```

```python
tips.columns = ['總票價', '小費', '吸煙者', '日期', '時間', '人數']
tips
```

```python
tips.groupby('吸煙者')[['人數']].sum()
```

```python
tips.groupby('吸煙者')[['人數','總票價']].sum()
```

```python
result = tips.groupby(['吸煙者','日期'])[['人數','總票價']].sum()
result
```

```python
tips
grouped = tips.groupby(['吸煙者','日期'])[['小費']]
grouped.mean()
```
