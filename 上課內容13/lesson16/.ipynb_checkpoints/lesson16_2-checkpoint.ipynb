```python
import numpy as np
import pandas as pd
dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
```

```python
dataFrame.iloc[-20:]
```

```python
newDataFrame = dataFrame.replace([0,'… '],np.nan)
newDataFrame.iloc[-20:]
```

```python
dropDataFrame=newDataFrame.dropna()
dropDataFrame.columns = dropDataFrame.iloc[0]
dropDataFrame=dropDataFrame.drop(index=0)
```

```python
dropDataFrame
```

```python
dropDataFrame.dtypes
```

```python
dropDataFrame['年底人口數'] = dropDataFrame['年底人口數'].astype('int32')
dropDataFrame.dtypes
dropDataFrame['土地面積'] = dropDataFrame['土地面積'].astype('float')
dropDataFrame.dtypes
dropDataFrame['人口密度'] = dropDataFrame['人口密度'].astype('float')
dropDataFrame.dtypes
```
