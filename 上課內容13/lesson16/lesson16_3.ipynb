```python
import matplotlib.pyplot as plt
import numpy as np
data = np.arange(10)
plt.figure()
plt.plot(data)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.plot(data)

ax2 = fig.add_subplot(2,2,2)
ax2.plot(np.random.randn(50))


ax3 = fig.add_subplot(2,2,3)
ax3.plot(np.random.randn(50).cumsum())

plt.savefig('figpath.png')
```
