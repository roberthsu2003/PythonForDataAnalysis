```python
import matplotlib.pyplot as plt
import numpy as np
data = np.arange(10)
plt.figure()
plt.plot(data)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.plot(data)

ax2 = fig.add_subplot(2,2,2)
ax2.plot(np.random.randn(50))


ax3 = fig.add_subplot(2,2,3)
ax3.plot(np.random.randn(50).cumsum())

plt.savefig('figpath.png')
```

```python
import pandas as pd
series = pd.Series(np.abs(np.random.randn(5)).cumsum(),index=['a1','a2','a3','a4','a5'])
print(series)
#series->plot()
series.plot()
```

```python
dataFrame=pd.DataFrame(np.random.randn(5,4),index=['b1','b2','b3','b4','b5'],columns=['A','B','C','D'])
dataFrame.plot()
plt.savefig('figpath.png')
```
